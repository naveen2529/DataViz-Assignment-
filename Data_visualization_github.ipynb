```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the CSV file
file_path = "indian_food.csv"
df = pd.read_csv(file_path)

# Display the first few rows
df.head()
```

### a)	In the state column, replace the -1 with the mode of state column.

```python
# Find the mode (most frequent value) of the 'state' column
mode_state = df['state'].mode()[0]

# Replace -1 with the mode value
df['state'] = df['state'].replace('-1', mode_state)

# Save the cleaned dataset (optional)
df.to_csv("cleaned_indian_food.csv", index=False)

# Display the first few rows to verify
print(df.head())
df.head()
```

### b)	Whatever the state you got in previous question, fill the corresponding region in the region column.

```python
# Find the corresponding region for Gujarat
gujarat_region = df[df['state'] == 'Gujarat']['region'].mode()[0]

# Replace -1 in the 'region' column with Gujarat's region
df['region'] = df['region'].replace('-1', gujarat_region)

# Save the cleaned dataset (optional)
df.to_csv("cleaned_indian_food.csv", index=False)

# Display the first few rows to verify
print(df.head())
```

### c)	Replace the -1 with mean for cook time variable.

```python
# Replace -1 in 'cook_time' with the mean (ignoring -1 values)
mean_cook_time = df[df['cook_time'] != -1]['cook_time'].mean()
df['cook_time'] = df['cook_time'].replace(-1, round(mean_cook_time))

# Save the cleaned dataset (optional)
df.to_csv("cleaned_indian_food.csv", index=False)

# Display the first few rows to verify
print(df.head())
```

### d)	Most preferred flavours by number of customers using pie chart.

```python

# Load the dataset
file_path = "cleaned_indian_food.csv"
df = pd.read_csv(file_path)

# Count occurrences of each flavor profile
flavor_counts = df['flavor_profile'].value_counts()

# Define a green color palette
colors = ['#006400', '#228B22', '#32CD32', '#7CFC00', '#ADFF2F']

# Plot pie chart with green shades
plt.figure(figsize=(8, 8))
plt.pie(flavor_counts, labels=flavor_counts.index, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title("Most Preferred Flavours by Number of Customers")
plt.show()
```

### e)	Most requested course by customers using Bar chart. Print the labels also on the top of each bar. Then sort the bar chart in descending order.

```python
# Count occurrences of each course and sort in descending order
course_counts = df['course'].value_counts().sort_values(ascending=False)

# Plot bar chart
plt.figure(figsize=(8, 6))
bars = plt.bar(course_counts.index, course_counts.values, color='#32CD32')

# Add labels on top of each bar
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), str(bar.get_height()), 
             ha='center', va='bottom', fontsize=12, fontweight='bold')

# Set title and labels
plt.xlabel("Course")
plt.ylabel("Number of Requests")
plt.title("Most Requested Course by Customers")
plt.xticks(rotation=45)

# Show plot
plt.show()
```

### f)	Comparison between Vegetarians and Non-Veg orders in each region. Hint: You can use stacked/grouped bar chart for comparison. 

```python
# Count the number of vegetarian and non-vegetarian dishes per region
diet_region_counts = df.groupby(['region', 'Diet']).size().unstack()

# Plot grouped bar chart
plt.figure(figsize=(10, 6))
diet_region_counts.plot(kind='bar', stacked=False, colormap='Greens', edgecolor='black', width=0.7)

# Set title and labels
plt.xlabel("Region")
plt.ylabel("Number of Dishes")
plt.title("Comparison of Vegetarian and Non-Vegetarian Orders in Each Region")
plt.xticks(rotation=45)
plt.legend(title="Diet")

# Show plot
plt.show()
```

### Stacked Chart

```python
# Plot stacked bar chart
plt.figure(figsize=(10, 6))
diet_region_counts.plot(kind='bar', stacked=True, colormap='Greens', edgecolor='black', width=0.7)

# Set title and labels
plt.xlabel("Region")
plt.ylabel("Number of Dishes")
plt.title("Comparison of Vegetarian and Non-Vegetarian Orders in Each Region (Stacked)")
plt.xticks(rotation=45)
plt.legend(title="Diet")

# Show plot
plt.show()
```

### g)	Create the box plot for Order cooking time to see the distribution. Also check if there are any outliers in this column.

```python
# Create a box plot for cook_time distribution
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['cook_time'], color='green')

# Set title and labels
plt.xlabel("Cooking Time (minutes)")
plt.title("Box Plot of Order Cooking Time")

# Show plot
plt.show()

# Identify outliers using the IQR method
Q1 = df['cook_time'].quantile(0.25)
Q3 = df['cook_time'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = df[(df['cook_time'] < lower_bound) | (df['cook_time'] > upper_bound)]

# Display number of outliers and their details
print(f"Number of outliers: {outliers.shape[0]}")
print(outliers[['Name', 'cook_time']])
```

### 3.	Name any 5 types of charts you are familiar with (using any of 3 libraries).Write the sample code also to create those charts using random data.


#### Bar chart for random data in metplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(10, 100, size=len(categories))  # Random values between 10 and 100

# Create bar chart
plt.figure(figsize=(8, 5))
plt.bar(categories, values, color='green', alpha=0.7)

# Add labels and title
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Random Data Bar Chart")

# Show values on top of bars
for i, v in enumerate(values):
    plt.text(i, v + 2, str(v), ha='center', fontsize=12)

# Show the plot
plt.show()
```

#### Pie Chart for random data in metplotlib

```python
# Generate random data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(10, 100, size=len(categories))  # Random values between 10 and 100

# Create pie chart
plt.figure(figsize=(7, 7))
plt.pie(values, labels=categories, autopct='%1.1f%%', colors=plt.cm.Greens(np.linspace(0.3, 0.9, len(categories))), startangle=140)

# Title
plt.title("Random Data Pie Chart")

# Show the plot
plt.show()
```

#### Line Chart for random data in metplotlib

```python
# Generate random data
x = np.arange(1, 11)  # X-axis values (1 to 10)
y = np.random.randint(10, 100, size=len(x))  # Random Y values between 10 and 100

# Create line chart
plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o', linestyle='-', color='green', markersize=8, linewidth=2, alpha=0.8)

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Random Data Line Chart")

# Show values on points
for i, v in zip(x, y):
    plt.text(i, v + 2, str(v), ha='center', fontsize=10)

# Show the plot
plt.grid(True)  # Add grid for better readability
plt.show()
```

#### Scatter Chart for random data in metplotlib

```python
# Generate random data
x = np.random.randint(1, 100, 50)  # 50 random X values between 1 and 100
y = np.random.randint(1, 100, 50)  # 50 random Y values between 1 and 100
sizes = np.random.randint(20, 200, 50)  # Random sizes for scatter points
colors = np.random.rand(50)  # Random colors for each point

# Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, s=sizes, c=colors, cmap='Greens', alpha=0.7, edgecolors='black')

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Random Data Scatter Plot")

# Show grid
plt.grid(True, linestyle="--", alpha=0.5)

# Show the plot
plt.show()
```
